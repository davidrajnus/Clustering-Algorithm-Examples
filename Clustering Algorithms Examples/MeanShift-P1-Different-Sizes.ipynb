```python
# Shift Mean Clustering
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset
dataset = pd.read_csv('Data/Aggregation.csv')
data = dataset.iloc[:, [0, 1]].values
```

```python
# visalizing the dataset
plt.scatter(data[:, 0], data[:, 1], s = 10, c = 'black')
```

```python
# Using the elbow method to find the optimal number of clusters
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++',max_iter = 300, n_init = 10)
    kmeans.fit(data)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Fitting Shift Mean to the dataset
from sklearn.cluster import MeanShift, estimate_bandwidth

bandwidth = estimate_bandwidth(data, quantile=0.085)
meanshift = MeanShift(bandwidth=bandwidth)
labels = meanshift.fit_predict(data)
np.unique(labels)
```

```python
# Visualising the clusters
plt.scatter(data[labels == 0, 0], data[labels == 0, 1], s = 10, c = 'blue')
plt.scatter(data[labels == 1, 0], data[labels == 1, 1], s = 10, c = 'red')
plt.scatter(data[labels == 2, 0], data[labels == 2, 1], s = 10, c = 'green')
plt.scatter(data[labels == 3, 0], data[labels == 3, 1], s = 10, c = 'brown')
plt.scatter(data[labels == 4, 0], data[labels == 4, 1], s = 10, c = 'pink')

plt.scatter(data[labels == 5, 0], data[labels == 5, 1], s = 10, c = 'silver')
plt.scatter(data[labels == 6, 0], data[labels == 6, 1], s = 10, c = 'magenta')
plt.scatter(data[labels == 7, 0], data[labels == 7, 1], s = 10, c = 'orange')
plt.scatter(data[labels == 8, 0], data[labels == 8, 1], s = 10, c = 'Yellow')

plt.title('Clusters results on the datasets')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.show()
```
