```python
# DBSCAN Clustering
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset
dataset = pd.read_csv('Data/Mall_Customers.csv')               
data = dataset.iloc[:, [3, 4]].values
data.shape
```

```python
# visalizing the dataset
plt.scatter(data[:, 0], data[:, 1], s = 10, c = 'black')
```

```python
# Fitting DBSCAN to the dataset
import hdbscan
hdbscan = hdbscan.HDBSCAN(min_cluster_size=10, min_samples = 5)
labels = hdbscan.fit_predict(data)
hdbscan.condensed_tree_.plot(select_clusters=True)
np.unique(labels)

#hdbscan.single_linkage_tree_.plot()
```

```python
# Visualising the clusters
plt.scatter(data[labels == -1, 0], data[labels == -1, 1], s = 10, c = 'black') # this line has been added 

plt.scatter(data[labels == 0, 0], data[labels == 0, 1], s = 10, c = 'blue')
plt.scatter(data[labels == 1, 0], data[labels == 1, 1], s = 10, c = 'red')
plt.scatter(data[labels == 2, 0], data[labels == 2, 1], s = 10, c = 'green')
plt.scatter(data[labels == 3, 0], data[labels == 3, 1], s = 10, c = 'brown')
plt.scatter(data[labels == 4, 0], data[labels == 4, 1], s = 10, c = 'grey')

plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```
